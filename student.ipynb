### Internal classes and functions for Data Scrubbing

```python
# Special functions for collecting NaN unique values
def define_nans(df_list):
    '''
    extract any values which can't be converted
    to numerical format to separate list
    '''
    df_list = list(df_list)
    res = []
    for i in df_list:
        try:
            i = float(i)
        except Exception:
            res.append(i)
        else:
            pass
        finally:
            res = list(set(res))
    return res

# Class for raw data analysing
class LookRawColGeneral:
    def __init__(self,df,col_name):
        self.df = df
        self.col_name = col_name
    
    def show_unique_values(self):
        _res = dict()
        for i in self.df[self.col_name]:
            if i in _res:
                _res[i]+=1
            else:
                _res[i]=1
        _name = '# of '+str(self.col_name)
        _res = pd.Series(data=_res, index=_res.keys())
        _res = pd.DataFrame(_res)
        _res = _res.rename(columns = {0: _name})
        return _res

    def show_var_stats(self):
        return self.df[self.col_name].describe()

def smart_table(df_raw):
    df_overview = pd.DataFrame([],index=df_raw.columns)
    df_overview['datatypes'] = df_raw.dtypes
    df_overview['nan_count'] = list(df_raw.isna().sum())

    _unique_val_count = []
    for i in df_raw.columns:
        _unique_val_count.append(len(df_raw[i].unique()))

    df_overview['uniq_val_count'] = _unique_val_count
    df_overview['min_val'] = df_raw.min()
    df_overview['max_val'] = df_raw.max()
    df_overview['median_val'] = df_raw.median()
    df_overview['q5%'] = df_raw.quantile(0.05)
    df_overview['q95%'] = df_raw.quantile(0.95)

    _negative_val_count = []
    for i in df_raw.columns:
        if df_raw[i].dtypes!='object':
            _negative_val_count.append(len(df_raw[i].loc[df_raw[i]<0]))
        else:
            _negative_val_count.append('N-A')
    df_overview['negative_val_count'] = _negative_val_count

    _zero_val_count = []
    for i in df_raw.columns:
        if df_raw[i].dtypes!='object':
            _zero_val_count.append(len(df_raw[i].loc[df_raw[i]==0]))
        else:
            _zero_val_count.append('N-A')
    df_overview['zero_val_count'] = _zero_val_count
    return df_overview   
```

#### Internal classes and functions

Before going to the next steps, let me briefly provide descriptions for created classes and functions in cell bellow

  - "convert_to_dummy" function - transform given column by using 1-hot-spot
    - if optional bin_list given, the function create binned category variable
    - for both cases (with and without binning) initial column dropped from df
  - "df_4_test" function - is a simple function, which concatenate target column ('price') and given columns as predictors
    - I've created this function for fast adding of dummy-subsets
  - "find_cat+group" function - create a list of dummy-variables within 1 subgroups
    - results of this function is used for df_4_test
  - "RegTest" class - is a class for making linear regression with statsmodels.OLS
    - class.show_summary() - print standart ols summary
    - class.get_Rsq() - return value of R-squared-adjusted
    - class.get_coef_pvals() - return a dataframe with coef-values, their p-values and bool-type column, where True value for a coefficient means, that corresponding p-value higher than selected alpha-level

```python
# support functions and classes for following analysis
import statsmodels.api as sm

def convert_to_dummy(df, col, bin_list = None):
    if bin_list != None:
        df[col] = pd.cut(df[col], bin_list)
    else:    
        df[col] = df[col].astype('category')
    x_dummy = pd.get_dummies(df[col],prefix=col)
    df.drop(columns=col, axis=1, inplace=True)
    df = pd.concat([df, x_dummy], axis=1)
    return df

def df_4_test(df, col_list, target_col):
    df = pd.concat([df[target_col], df[col_list]], axis=1)
    return df

def find_cat_group(df, shablon):
    res = []
    col_name = list(df.columns)
    for i in col_name:
        if str(i).startswith(shablon) or str(i) == shablon:
            res.append(i)
    return res

class RegTest:
    def __init__(self,df,target_col, alpha_level=0.05):
        self.df = df
        self.target_col = target_col
        self.alpha_level = alpha_level
        self.model = self._run_reg()
    
    def _run_reg(self):
        self.predictors = sm.add_constant(self.df.drop(self.target_col, axis=1))
        self.model = sm.OLS(self.df[self.target_col], self.predictors).fit()
        return self.model
    
    def show_summary(self):
        print(self.model.summary())
    
    def get_Rsq(self):
        return self.model.rsquared_adj
    
    def get_coef_pvals(self):
        self.res = pd.DataFrame([])
        self.res['coefs'] = self.model.params
        self.res['PVals'] = self.model.pvalues
        _list = []
        for i in self.res['PVals']:
            if i>=self.alpha_level:
                _list.append(True)
            else:
                _list.append(False)
        self.res['T-stat-fail'] = _list
        return self.res

df_cutted = df_raw.copy() # checkpoint for next steps
```

```python
# dataset preparation
from scipy.stats import skew
cont_vars = []
for i,j in zip(df_cutted.dtypes[1:], df_cutted.columns[1:]):
    if i != 'uint8':
        cont_vars.append(j)
df_unscaled = df_cutted[cont_vars] # checkpoint here
alpha_level1 = 0.5
alpha_level2 = 1

# transformation functions
def var_to_lg(df, var):
    var_log = pd.DataFrame([])
    var_log_name = 'log_'+str(var)
    var_log[var_log_name] = np.log(df[var])
    return var_log

def min_max_scale(df, var):
    var_scaled = pd.DataFrame([])
    var_scaled_name = 'scaled_'+str(var)
    var_scaled[var_scaled_name] = (df[var] - df[var].min())/(df[var].max() - df[var].min())
    return var_scaled

def make_plot(df1, df2, var1, var2):
    fig, ax = plt.subplots(1,2,figsize=(10,3))
    sk1, sk2 = round(skew(df1[var1]),2), round(skew(df2[var2]),2)
    ax[0].set_title('skew: '+str(sk1))
    ax[1].set_title('skew: '+str(sk2))
    sns.distplot(df1[var1],hist=False,color='r',label='original',ax = ax[0])
    sns.distplot(df2[var2],hist=False,color='g',label='logged',ax = ax[1])
    return None
```

## Model fitting

### description of model fitting algorythm

The Model fitting step contains of 4 main steps:
  1. Features selection.
  2. Model fitting
  3. Model validation (testing of residuals)
  4. Transforming final coefficients for report

1. Feature selection. 
   - chosen method - RFE (with range from 5 to 60 with step 1)
   - main perfomance parameters for each x-set of features are Rsq-adj and MSE
   - for MSE calculation in each iteration (for x-set) I will run K-fold block (with k = 0.2)
   - 2-3 cases will be selected for next step
     - best by Rsq
     - best by MSE
     - balanced case
   - 2 cases instead of 3 cases will be taken in a difference between best-by-Rqs and best-by-MSE will be small

2. Testing p-values
    - in this step, I will check on P-values and adjust features for each scenario
    
3. Model fitting.
   - I will split dataset to Train and Test subsets for calculating coefficients
   - As mentioned in the previous part, all calculations will be done for 2 or 3 selected feature-sets
   - Split procedure will be realized in cycles from 10% to 30% with 5%-step (5 iterations in total)
   - For each x-sets of features, the optimal split will be selected. Selection criteria - MSE from test subset


3. Model validation.
   - Based on 3 coefficient valuation results 3 subsets of residuals will be generated
   - Residuals vectors will be tested on Normality (by visualization)
   - Based on this analysis optimal case will be chosen


4. Transforming final coefficients
   - In this step, all transformations on variables made before (logging, shifting and normalizing) will be applied (reverse) on coefficients for preparing final formula.

```python
# preparing dataset
Y = df_cutted['log_price'] 
Y = pd.concat([Y], axis=1) # check how it works with SKLEARN in form of DF instead of DS
X = df_cutted.drop(['log_price'], axis=1)
CV_5 = 5

# importing modules
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
from scipy.stats import jarque_bera
from scipy.stats import kurtosis

# creating internal functions
def calc_Rsq_adj(Y,Y_hat,k):
    Y_np = Y.values
    Y_mu = Y_np.mean()
    tss_array = (Y_np - Y_mu)**2
    rss_array = (Y_hat - Y_mu)**2
    Rsq_ = rss_array.sum()/tss_array.sum()
    Rsq_a = 1 - (1-Rsq_)*((len(tss_array)-1)/(len(tss_array)-k))
    return Rsq_a

def calc_Y_hat(X, Y, coefs):
    _Y_hat = pd.DataFrame([])
    target_name_ = 'y_predict'
    if type(coefs) == float or type(coefs) == int:
        coefs = [coefs]
        _Y_hat[target_name_] = list(coefs[0] for i in range(len(X)))
    else:
        _Y_hat[target_name_] = sum(list(coefs[j+1]*X.iloc[:,j] for j in range(len(coefs)-1))) + coefs[0]
    return _Y_hat   
```

### Feature selection

```python
# RFE - running code here - take up 30 sec - please wait
results_feat = dict()
results_rsq = dict()
results_mse = dict()
k = KFold(n_splits=5, shuffle=True, random_state=1)
predictors = X

for i in range(5,61):
    # 1. initialisation
    linreg = LinearRegression()
    selector = RFE(linreg, n_features_to_select=i)
    selector = selector.fit(predictors, Y.iloc[:,0]) 
    estimators = selector.estimator_
    selected_cols = predictors.columns[selector.support_]
    results_feat[i] = selected_cols # add to dict
    # 2. calc
    linreg.fit(predictors[selected_cols],Y.iloc[:,0])
    y_hat = linreg.predict(predictors[selected_cols])
    # 3. R-adj
    rsq = calc_Rsq_adj(Y,y_hat,i)
    results_rsq[i] = rsq
    # 4. K-fold
    cv_5 = np.mean(cross_val_score(linreg,X[selected_cols], Y, cv=k,scoring='neg_mean_squared_error'))
    results_mse[i] = cv_5
```

```python
# creating resulting table
df_kf_results = pd.DataFrame(index = results_rsq.keys())
df_kf_results['features'] = results_feat.values()
df_kf_results['mse'] = results_mse.values()
df_kf_results['R_sq_adj'] = results_rsq.values()

# discovering RFE results
best_feature_by_mse = df_kf_results.sort_values(by='mse', ascending=False)[:1]
best_feature_by_Rsq = df_kf_results.sort_values(by='R_sq_adj', ascending=False)[:1]
print('scenario 1: best by Rsq')
print(50*'-')
display(best_feature_by_Rsq)
print('scenario 2: best by MSE')
print(50*'-')
display(best_feature_by_mse)

fig,ax = plt.subplots(1,2, figsize=(14,6))
ax[0].set_title('MSE curve, in negative form')
ax[0].set_xlabel('# of features')
ax[1].set_title('Rsq.adj curve')
ax[1].set_xlabel('# of features')
ax[0].plot(df_kf_results.index, df_kf_results.mse, c='r')
ax[1].plot(df_kf_results.index, df_kf_results.R_sq_adj, c='g');
```

***Before transforming back coefficients, let's looks, how to simplify outcome formula by using standart math rules.***
1. Firstly, assume that we have:
  - x1 and x2 predictors, which were logged and after scaled, so **z1 = ( ln(x1) - min(ln(x1)) )/(max(ln(x1))-min(ln(x1))**
  - here and bellow, for simplicity, I will write just min1 and min2 instead of full formula
  - x3 and x4, which were only scaled, so **z3 = (x3-min3) / (max3-min3)**
  - x5 and x6, which were not changed, so **z5 = x5**


2. As one of the main outcome for the project is to issue a predictive formula for target
  - In this project, target var was logged: **y = ln(target)**


3. We have 3 way of showing results: 
  - in form of **ln(target) = f(z1,..,z6)** and ofcourse gives all details to User of how to transform original x to z
  - in form of **ln(target) = f(x1,..,x6)**, so, only the 1 simple operation is required from User (to transform y). In this project I will show in resulting table this version of outcome as final formula
  - in form of **target = f(x1,..,x6)**, which I think the best way, since it shows true nature of target dependency from x (could be not linear)


4. Here the main initial formulas I used for simplification:
  - $\left(a\right)\log_e{y}\equiv\ln{\left(y\right)}=z,\ e^z=y$
  - $\left(b\right){\ y}^{\left(a+b\right)}=y^ay^b$
  - $\left(c\right)e^{\ln{\left(y\right)}}=y$
  - $(d)\ln{\left(yz\right)}=ln\left(y\right)ln\left(z\right)$
  - $(e)\ ln\left(y^k\right)=kln\left(y\right)$


5. Formula simplification


- original formula: 

$\ln{\left(y\right)}={C+\alpha}_1z_1+\alpha_2z_2+\alpha_3z_3+\alpha_4z_4+\alpha_5z_5+\alpha_6z_6$
  
  
- let's transform it in terms of x: 

$\ln{\left(y\right)}=C+\alpha_1\frac{ln\left(x_1\right)-\min_1}{\max_1-\min_1}+\alpha_2\frac{ln\left(x_2\right)-\min_2}{\max_2-\min_2}+\alpha_3\frac{x_3-\min_3}{\max_3-\min_3}+\alpha_4\frac{x_4-\min_4}{\max_4-\min_4}+\alpha_5x_5{+\alpha}_6x_6$


- looking to x1 and x2: 

$\alpha_i\frac{ln\left(x_i\right)-\min_i}{\max_i-\min_i}=\left(\frac{\alpha_i}{\max_i-\min_i}\right)ln\left(x_i\right)-\frac{\alpha_i\min_i}{\max_i-\min_i}\equiv\beta_iln\left(x_i\right)-\gamma_i$


- same is for x3 and x4, but without ln: 

$\alpha_i\frac{x_i-\min_i}{\max_i-\min_i}=\left(\frac{\alpha_i}{\max_i-\min_i}\right)x_i-\frac{\alpha_i\min_i}{\max_i-\min_i}\equiv\beta_ix_i+\gamma_i$


- so, regression formula will be in form of: 

$\ln{\left(target\right)}=\beta_1ln\left(x_1\right)+\beta_2ln\left(x_2\right)+\beta_3x_3+\beta_4x_4+\alpha_5x_5+\alpha_6x_{6+}\left[C\ -\ \gamma_1-\gamma_2{-\gamma}_3{-\gamma}_4\right]$


- from the formula above, we can easily recalculate alpha-coefs to beta-coefs and new Intercept value for User, and only what User needs is to transform some of features and target by logging.


- Now, let's try take out all logs from formula:

${target=e}^{ln\left(target\right)}=e^{\left(\beta_1ln\left(x_1\right)+\beta_2ln\left(x_2\right)\right)+\left(\beta_3x_3+\beta_4x_4\right)+\left(\alpha_5x_5+\alpha_6x_6\right)+D}$


- By using main formulas final equation will be in next form:

$target=\ \left(x_1^{\beta_1}x_2^{\beta_2}\right)\bullet\left(e^{\left(x_3+x_4+x_5+x_6\right)}\right)\bullet\left(e^{\left(\beta_3+\beta_4+\alpha_5+\alpha_6\right)}\right)\bullet\left(e^D\right)=CONST\bullet\left(x_1^{\beta_1}x_2^{\beta_2}\right)\bullet\left(e^{\left(x_3+x_4+x_5+x_6\right)}\right)$

- so, in formula above, we knows all alpha and C, and we just need to convert it to beta and CONST

- also, we see the type of functional dependency from each predictor by looking to this new form


```python
# 1. splitting original coefficients into 3 groups and adding alpha-coefs
scaled_log_alpha = dict().fromkeys(['scaled_log_sqft_living','scaled_log_sqft_living15','scaled_log_sqft_lot'])
scaled_alpha = dict().fromkeys(['scaled_yr_built','scaled_lat'])
dummy_alpha = dict().fromkeys(list(final_coefs.keys())[6:])

for key in scaled_log_alpha.keys():
    scaled_log_alpha[key]=[final_coefs[key],None,None]

for key in scaled_alpha.keys():
    scaled_alpha[key]=[final_coefs[key],None,None]

for key in dummy_alpha.keys():
    dummy_alpha[key]=final_coefs[key]

# 2. Obtaining beta and gamma
# because items' count to be transormed to big, I didn't write automatized code here
scaled_log_alpha['scaled_log_sqft_living'][1] = df_param['var_min'][0]
scaled_log_alpha['scaled_log_sqft_living'][2] = df_param['var_max'][0]
scaled_log_alpha['scaled_log_sqft_living15'][1] = df_param['var_min'][5]
scaled_log_alpha['scaled_log_sqft_living15'][2] = df_param['var_max'][5]
scaled_log_alpha['scaled_log_sqft_lot'][1] = df_param['var_min'][1]
scaled_log_alpha['scaled_log_sqft_lot'][2] = df_param['var_max'][1]
scaled_alpha['scaled_yr_built'][1] = df_param['var_min'][2]
scaled_alpha['scaled_yr_built'][2] = df_param['var_max'][2]
scaled_alpha['scaled_lat'][1] = df_param['var_min'][3]
scaled_alpha['scaled_lat'][2] = df_param['var_max'][3]

scaled_log_bg = dict().fromkeys(scaled_log_alpha.keys())
scaled_bg = dict().fromkeys(scaled_alpha.keys())

for k in scaled_log_bg.keys():
    _min = scaled_log_alpha[k][1]
    _max = scaled_log_alpha[k][2]
    _alpha = scaled_log_alpha[k][0]
    _beta = _alpha/(_max - _min)
    _gamma = _beta*_min
    scaled_log_bg[k] = [_beta, _gamma]
    
for k in scaled_bg.keys():
    _min = scaled_alpha[k][1]
    _max = scaled_alpha[k][2]
    _alpha = scaled_alpha[k][0]
    _beta = _alpha/(_max - _min)
    _gamma = _beta*_min
    scaled_bg[k] = [_beta, _gamma]

# 3 Calculating new coefficients
i = 0
for value in scaled_log_bg.values():
    i = i +value[1]
for value in scaled_bg.values():
    i = i +value[1]
D = final_coefs['const'] - i

# 4. Logged version
original_feat_list = []
transform_list = []
transformed_coef_list = []
final_coef_list = []
comment_list = []

for idx,i in enumerate(final_coefs.keys()):
    transformed_coef_list.append(final_coefs[i])
    if idx == 0 or idx>=6:
        original_feat_list.append(i)
        transform_list.append('-')
        if idx == 0:
            final_coef_list.append(D)
            comment_list.append('')
        else:
            final_coef_list.append(final_coefs[i])
            comment_list.append(f'x {i}')
    elif idx == 3 or idx == 4:
        original_feat_list.append(i[7:])
        transform_list.append('scaled')
        final_coef_list.append(scaled_bg[i][1])
        comment_list.append(f'x {i[7:]}')
    else:
        original_feat_list.append(i[11:])
        transform_list.append('logged and scaled')
        final_coef_list.append(scaled_log_bg[i][1])
        comment_list.append(f'x ln({i[11:]})')

original_feat_list.insert(0, 'price (target)')
transform_list.insert(0, 'log_price')
transformed_coef_list.insert(0, ' ')
final_coef_list.insert(0, ' ')
comment_list.insert(0, 'ln(price) =')



outcome_table = pd.DataFrame(index=original_feat_list)
outcome_table['coefs from linreg'] = transformed_coef_list
outcome_table['transforming'] = transform_list
outcome_table['re-scaled coefs'] = final_coef_list
outcome_table['formula'] = comment_list

final_formula = ''
for i,j in zip(final_coef_list,comment_list):
    if type(i) == str:
        final_formula = final_formula + j
    else:
        if i < 0:
            final_formula = final_formula + ' ' + str(round(i,2)) +  j
        else:
            final_formula = final_formula + ' + ' + str(round(i,2)) + j

# 5. Now, let's look to the FINAL FORMULA
print('regression formula')
print('='*120)
print(final_formula) 
print('-'*120)
```

**Examples of interpretation of modelling results**
  - Because unit and scale for different feauters are hard to compare with each other, I didn't show here some kind or ranking. 
  - If we choose 1 of features and assume to fix other features with some values, we can look to formula above as the regression with 1 predictor - for illustrative purposes. 
  - Let's select 3 predictors: sqft_living, lat, waterfront, which are from different subgroups of features, and treat the model by choosing 1 of them and fix all other features. And let's assume, that all other features (with some fixed variables) + const = A (another_const).
  - Selected var = sqft_living. ln(Price_forecast) = A + 2.21 x ln(Sqft_Living) => Price_forecast = (EXP^A) x (2.21 x Sqft_Living).-> Price Elasticity on Sqft_living = delta(price)/delta(sqft_living) = (EXP^A) x 2.21. So, this change even for fixed values for another features, not behave like classical linear function, where dY/dX = fixed coef.
  - Selected var = lat. According to model results, the more houses located in North - the higher prices will be. Delta(Price) = Price x [EXP^(64xDelta(lat)) - 1].
  - Selected var = waterfront. It is bool-type variable with two possible values: 0 and 1. So, it gives: ln(Price with Watefront / Price without Waterfront) = 0.52 => Price with WaterFront = 1.68 x Price without Waterfront.
